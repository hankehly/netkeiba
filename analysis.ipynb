```python
import pandas as pd

import json

def load_data() -> pd.DataFrame:
    df_lines = pd.read_json('output/2018-10-23T13:28:27.jl', lines=True)
    json_dict = json.loads(df_lines.to_json(orient='records'))
    return pd.io.json.json_normalize(json_dict, sep='_')
```

```python
df = load_data()
```
